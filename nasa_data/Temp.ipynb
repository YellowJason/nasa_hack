```python
import numpy as np
import matplotlib.pyplot as plt
year = np.loadtxt(r'graph.txt', skiprows=5, usecols=0)
tmp = np.loadtxt(r'graph.txt', skiprows=5, usecols=1)
tmp_L = np.loadtxt(r'graph.txt', skiprows=5, usecols=2)

plt.figure(figsize=(8,6))
plt.plot(year, tmp, label='yearly data')
plt.plot(year, tmp_L, linewidth=3, label='trend')
plt.legend()
plt.xlabel('YEAR')
plt.ylabel('Temperature[$\degree$C]')
plt.title('Land-Ocean Temperature Index', fontsize=14)
plt.show()
```

```python
f = open("Temp_x.txt", 'w')
for i in range(len(year)):
    f.write(f"{str(int(year[i]))},\n")
f.close()
```

```python
f = open("Temp.txt", 'w')
for i in range(len(tmp)):
    f.write(f"{str(tmp[i])},\n")
f.close()
```
