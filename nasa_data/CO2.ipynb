```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
co2_x = np.loadtxt(r'co2_mm_mlo.txt',  skiprows=30, dtype = float, usecols = 2)
co2 = np.loadtxt(r'co2_mm_mlo.txt',  skiprows=30, dtype = float, usecols = 3)
```

```python
co2_mean = []
for i in range(len(co2)):
    sum = 0
    if (i+11) < len(co2):
        for j in range(12):
            sum += co2[i+j]
        co2_mean.append(sum/12)
    else:
        for j in range(12):
            sum += co2[i-j]
        co2_mean.append(sum/12)
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "JUL", "Aug", "Sep", "Oct", "Nov", "Dec"]

f = open("co2_x.txt", 'w')
for i in range(len(co2_x)):
    year = int(co2_x[i])
    month = co2_x[i]-year
    month = int(month*12)
    f.write(f"\"{months[month]},{str(year)}\",\n")
        
f.close()
```

```python
f = open("co2.txt", 'w')
for i in range(len(co2)):
    f.write(f"{str(co2[i])},\n")
f.close()
```

```python
f = open("co2.txt", 'w')
for i in range(len(co2)):
    f.write(f"{str(co2[i])},\n")
f.close()
```

```python
f = open("co2_mean.txt", 'w')
for i in range(len(co2_mean)):
    f.write(f"{str(co2_mean[i])},\n")
f.close()
```

```python
plt.figure(figsize=(8,6))
plt.plot(co2_x, co2_mean, linewidth = 3, label='All year')
plt.plot(co2_x, co2, linewidth = 1, label='All year')
plt.show()
```
